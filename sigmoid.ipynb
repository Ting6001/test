```python
# =========== sigmoid圖 ==================
import math
import matplotlib.pyplot as plt

x = []
x = list(range(-10, 11)) # [-10,-9,...9,10]的陣列

def sigmoid(x):
    return 1/(1+math.exp(-x))

def linear(x):
  return x # y=x的函數

y_linear = [linear(x) for x in x]

sigmoid_x1 = [sigmoid(a) for a in x] # y是經過sigmold轉換過的值

sigmoid_x2 = [sigmoid(a*0.5) for a in x] # 進去sigmoid的值愈小，愈趨近於一條斜直線

sigmoid_x3 = [sigmoid(a*10) for a in x] # 進去sigmoid的值愈大，激活跳起的斜率愈大

sigmoid_1 = [sigmoid(a) for a in x] # 加上bias偏移量可以讓圖形左右移動

sigmoid_2 = [sigmoid(a-4) for a in x] # -會讓圖形往右偏移

sigmoid_3 = [sigmoid(a+4) for a in x] # +會讓圖形往左偏移

def plt_style(plt):
  ax = plt.gca()
  ax.spines['right'].set_color('none')
  ax.spines['top'].set_color('none')
  ax.xaxis.set_ticks_position('bottom')
  ax.spines['bottom'].set_position(('data',0))
  ax.yaxis.set_ticks_position('left')
  ax.spines['left'].set_position(('data',0))

plt.figure() # 做出一張畫布
# 畫第一個圖
plt.subplot(3, 1, 1) # (總共row, 總共col, 目前是要畫第幾張圖)
plt.plot(x, y_linear)
plt_style(plt)
# 畫第二個圖
plt.subplot(3, 1, 2)
plt.plot(x, sigmoid_x1)
plt.plot(x, sigmoid_x2)
plt.plot(x, sigmoid_x3)
plt_style(plt)
# 畫第三個圖
plt.subplot(3, 1, 3)
plt.plot(x, sigmoid_1)
plt.plot(x, sigmoid_2)
plt.plot(x, sigmoid_3)
plt_style(plt)

plt.show()
```
